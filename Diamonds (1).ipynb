```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_csv('diamonds.csv',lineterminator='\n')

df.shape

df.isnull().sum()

plt.figure (figsize =(20,8))
sns.histplot(df['price'],bins =50,kde=True)
plt.title("Distribution of Diamond Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['price'])
plt.title("Boxplot of Diamond Price")
plt.ylabel("Price")
plt.show()


plt.figure (figsize =(20,8))
sns.histplot(df['carat'],bins =50,kde=True)
plt.title("Distribution of Diamond carat")
plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['carat'])
plt.title("Boxplot of Diamond carat")
plt.ylabel("Carat")
plt.show()

plt.figure (figsize =(20,8))
sns.histplot(df['depth'],bins =50,kde=True)
plt.title("Distribution of Diamond depth")
plt.xlabel("depth")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['depth'])
plt.title("Boxplot of Diamond depth")
plt.ylabel("depth")
plt.show()

plt.figure (figsize =(20,8))
sns.histplot(df['table'],bins =50,kde=True)
plt.title("Distribution of Diamond table")
plt.xlabel("table")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['table'])
plt.title("Boxplot of Diamond table")
plt.ylabel("table")
plt.show()

plt.figure (figsize =(20,8))
sns.histplot(df['x'],bins =50,kde=True)
plt.title("Distribution of Diamond x")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['x'])
plt.title("Boxplot of Diamond x")
plt.ylabel("x")
plt.show()

plt.figure (figsize =(20,8))
sns.histplot(df['y'],bins =50,kde=True)
plt.title("Distribution of Diamond y")
plt.xlabel("y")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['y'])
plt.title("Boxplot of Diamond y")
plt.ylabel("y")
plt.show()

plt.figure (figsize =(20,8))
sns.histplot(df['z'],bins =50,kde=True)
plt.title("Distribution of Diamond z")
plt.xlabel("z")
plt.ylabel("Frequency")
plt.show()

plt.figure (figsize =(20,8))
sns.boxplot(y=df['z'])
plt.title("Boxplot of Diamond z")
plt.ylabel("z")
plt.show()
```
